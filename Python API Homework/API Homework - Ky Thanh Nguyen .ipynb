# Finding Random Cities: 

```python
# Dependencies: 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import random 
import requests 
import time 
from pprint import pprint 
from citipy import citipy 

# API keys: 
from config import api_key 

```

```python
# Find random latitude and longtitude: 
random_lat=[]
random_long=[]

while len(random_lat) < 1200:
    lat = np.random.uniform(-90, 90)
    long = np.random.uniform(-180,180)
    random_lat.append(lat)
    random_long.append(long)
    
df = pd.DataFrame ({"lat": random_lat, "long":random_long})
df.head()
```

```python
# Use Random latitude and longtitude to find city with country code: 
cities = []
countries = []

for index, row in df.iterrows(): 
    city_lat =(row["lat"])
    city_long = (row["long"])
    city_name = citipy.nearest_city(city_lat, city_long).city_name
    country_code = citipy.nearest_city(city_lat, city_long).country_code
    city_country_code = city_name + "," + country_code
    cities.append(city_country_code)
```

# Calling API Data 

```python
# Variables: 
cloudiness = []
humidity = []
temp = []
wind = []
lat = []
lng = []
new_cities = []
date = []
```

```python
record_counter = 0
set_counter = 0

# Starting URL for Weather Map API Call
print('------------------------') 
print('Beginning Data Retrieval')
print('------------------------')

# Orginial url: 
url = "https://api.openweathermap.org/data/2.5/weather?q="

# query url: 
for city in cities:
    query_url = url + city + "&appid=" + api_key
    response = requests.get(query_url).json()
  
    if record_counter < 50:
        record_counter += 1
    else:
        set_counter += 1
        record_counter = 0
# API Call to collect data into variables: 
    print('Processing record {} of set {} | {}'.format(record_counter, set_counter, city))
    print(query_url)
    try:
        cloudiness.append(response['clouds']['all'])
        humidity.append(response['main']['humidity'])
        temp.append(response['main']['temp_max'])
        wind.append(response['wind']['speed'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        new_cities.append(city)
        date.append(response['dt'])
       
    except:
        print("City not found!")
        pass

print('------------------------')
print('Data Retrieval Complete')
print('------------------------')
```

# Creating DataFrame and ScatterPlot: 

```python

df = pd.DataFrame ({"City with Country Code": new_cities,
                   "Latitude": lat,
                   "Max Temperature (F)": temp, 
                   "Windspeed": wind,
                   "Humidity": humidity,
                   "Cloudiness": cloudiness,
                   "Date": date})                    

df = df.head(500)
df.to_csv("df.csv", index=False, header=True)
df.head()
```

```python
# Scatter Plot showing the relationship of Latitude and Temperature: 
df.plot(kind="scatter", x= "Latitude", y= "Max Temperature (F)", grid=True, figsize=(8,4),
              title="Latitude vs Temperature")
plt.savefig("LatvsTemp.png")
plt.show()
```

```python
# Scatter Plot showing the relationship of Latitude and Humidity: 
df.plot(kind="scatter", x= "Latitude", y= "Humidity", grid=True, figsize=(8,4),
              title="Latitude vs Humidity")
plt.savefig("LatvsHumidity.png")
plt.show()
```

```python
# Scatter Plot showing the relationship of Latitude and Cloudiness: 
df.plot(kind="scatter", x= "Latitude", y= "Cloudiness", grid=True, figsize=(8,4),
              title="Latitude vs Cloudiness")
plt.savefig("LatvsCloudiness.png")
plt.show()
```

```python
# Scatter Plot showing the relationship of Latitude and Wind Speed: 
df.plot(kind="scatter", x= "Latitude", y= "Windspeed", grid=True, figsize=(8,4),
              title="Latitude vs Windspeed")
plt.savefig("LatvsWinds.png")
plt.show()
```
